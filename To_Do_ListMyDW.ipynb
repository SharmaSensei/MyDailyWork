```python
from datetime import datetime

# Initialize an empty list to store tasks
tasks = []

def show_tasks():
    """Display the current tasks with priority and due dates."""
    if not tasks:
        print("No tasks found.")
    else:
        print("\nCurrent Tasks:")
        for i, (task, priority, due_date) in enumerate(tasks, start=1):
            print(f"{i}. [{priority}] {task} (Due: {due_date})")

def add_task(task, priority, due_date):
    """Add a new task to the list with priority and due date."""
    tasks.append((task, priority, due_date))
    print(f'Added task: "{task}" with priority "{priority}" and due date "{due_date}"')

def remove_task(index):
    """Remove a task from the list by index."""
    if 0 <= index < len(tasks):
        removed_task = tasks.pop(index)
        print(f'Removed task: "{removed_task[0]}"')
    else:
        print("Invalid task number.")

def mark_complete(index):
    """Mark a task as complete."""
    if 0 <= index < len(tasks):
        print(f'Marked task as complete: "{tasks[index][0]}"')
        tasks[index] = (tasks[index][0] + " (Completed)", tasks[index][1], tasks[index][2])
    else:
        print("Invalid task number.")

# Main program loop
while True:
    print("\nTo-Do List Menu:")
    print("1. Show Tasks")
    print("2. Add Task")
    print("3. Remove Task")
    print("4. Mark Task Complete")
    print("5. Exit")

    choice = input("Choose an option (1-5): ")

    if choice == '1':
        show_tasks()
    elif choice == '2':
        task = input("Enter the task: ")
        priority = input("Enter the priority (High, Medium, Low): ")
        due_date = input("Enter the due date (YYYY-MM-DD): ")
        # Validate the due date format
        try:
            datetime.strptime(due_date, "%Y-%m-%d")
            add_task(task, priority, due_date)
        except ValueError:
            print("Invalid date format. Please use YYYY-MM-DD.")
    elif choice == '3':
        show_tasks()
        task_number = int(input("Enter the task number to remove: ")) - 1
        remove_task(task_number)
    elif choice == '4':
        show_tasks()
        task_number = int(input("Enter the task number to mark as complete: ")) - 1
        mark_complete(task_number)
    elif choice == '5':
        print("Exiting the To-Do List Application.")
        break
    else:
        print("Invalid choice. Please try again.")
```
